## Print all Palindrominc Partitions

```python




# Time: O(n*2^n)
# Aux: O(n^2)



# Utility function to check if string is palindrome
def isPalindrome(string, low, high):
    while low < high:
        if string[low] != string[high]:
            return False
        low += 1
        high -= 1
    return True


# Recursive Function to find all palindromic partitions of string
def allPalindromePartitions(allPart, currPart, start, n, string):
      
    if start >= n:
        
        # Make a copy of lst2 then append to lst1, because lists are passed by refference
        x = currPart.copy()

        allPart.append(x)

        return
    
    # Pick all possible ending points for substrings
    # Start = 0
    # 'n' = len(strint)
    for i in range(start, n):
        
        # If substring str[start...i] is palindrome
        if isPalindrome(string, start, i):
            
            # Add the substring to the result
            currPart.append(string[start: i + 1])
            
            # Recursion for remaining substring
            allPalindromePartitions(allPart, currPart, i + 1, n, string)
            
            # Remove substring str[start..i] from current partition
            currPart.pop()
    

# Function to print all possible palindromic partitions of the string
# Creates vectors and calls allPalindromicPartitions()
def allPalPartitions(string):
    
    n = len(string)
    
    # To store all palindromic partitions
    allPart = []
    
    # To store current palindromic partitions
    currPart = []
    
    # Call recursive function to generate all partitions and store in allPart
    allPalindromePartitions(allPart, currPart, 0, n, string)
    
    # Print all partitions generated from the above call
    for i in range(len(allPart)):
        for j in range(len(allPart[i])):
            print(allPart[i][j], end = ' ')
            print()
    
    
# Driver Code
if __name__ == "__main__":
    string = 'radar'
    allPalindromePartitions(string)
    
    
```

## Tower of Hanoi

```python
# I want to move a tower of size 'n'
# Assume we can move a stack of 'n-1' plates


def print_move(frum, to):
    print('move from ' + str(frum) + ' to ' + str(to))
    
    
def tower_of_hanoi(n, frum, to, spare):
    
    if n == 1:
        print_move(frum, to)
        print()
    
    else:
        

        tower_of_hanoi(n - 1, frum, spare, to)
        

        tower_of_hanoi(1, frum, to, spare)
       

        tower_of_hanoi(n - 1, spare, to, frum)
        
        

        
tower_of_hanoi(5, 1, 2, 3)
        
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def build_hanoi_tree(n, from_peg, to_peg, spare_peg, label_prefix="T", depth=0, G=None, parent=None, node_id_gen=[0]):
    if G is None:
        G = nx.DiGraph()

    node_id = f"{label_prefix}{node_id_gen[0]}"
    node_id_gen[0] += 1

    if n == 1:
        label = f"Move disk 1: {from_peg} → {to_peg}"
    else:
        label = f"T({n}): {from_peg} → {to_peg}"

    G.add_node(node_id, label=label)
    if parent:
        G.add_edge(parent, node_id)

    if n > 1:
        build_hanoi_tree(n - 1, from_peg, spare_peg, to_peg, label_prefix, depth + 1, G, node_id, node_id_gen)
        build_hanoi_tree(1, from_peg, to_peg, spare_peg, label_prefix, depth + 1, G, node_id, node_id_gen)
        build_hanoi_tree(n - 1, spare_peg, to_peg, from_peg, label_prefix, depth + 1, G, node_id, node_id_gen)

    return G

# Build tree for n = 3
tree = build_hanoi_tree(3, "1", "2", "3")

# Draw the tree
pos = nx.nx_agraph.graphviz_layout(tree, prog='dot')
labels = nx.get_node_attributes(tree, 'label')

plt.figure(figsize=(12, 8))
nx.draw(tree, pos, with_labels=False, arrows=True, node_color="lightyellow", node_size=3000, edgecolors='black')
nx.draw_networkx_labels(tree, pos, labels, font_size=9)

plt.title("Recursive Call Tree for tower_of_hanoi(3, 1, 2, 3)", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# Generate a labeled move trace showing the actual disk number at each step

def generate_explicit_disk_trace(n, from_peg, to_peg, spare_peg, depth=0, move_list=None):
    if move_list is None:
        move_list = []

    indent = "  " * depth  # visual indent for recursion depth

    if n == 1:
        move_list.append(f"{indent}Move disk 1 from {from_peg} to {to_peg}")
    else:
        # Step 1: Move n-1 disks to spare
        generate_explicit_disk_trace(n - 1, from_peg, spare_peg, to_peg, depth + 1, move_list)

        # Step 2: Move the actual largest disk (n)
        move_list.append(f"{indent}Move disk {n} from {from_peg} to {to_peg}")

        # Step 3: Move n-1 disks from spare to destination
        generate_explicit_disk_trace(n - 1, spare_peg, to_peg, from_peg, depth + 1, move_list)

    return move_list

# Generate trace
explicit_trace = generate_explicit_disk_trace(3, 1, 2, 3)

# Display the trace
import pandas as pd


df = pd.DataFrame({"Tower of Hanoi Moves (n=3) with Actual Disk Numbers": explicit_trace})

df
```

```python
import turtle


def draw_triangle(points, color, my_turtle):
    my_turtle.fillcolor(color)
    my_turtle.up()
    my_turtle.goto(points[0][0], points[0][1])
    my_turtle.down()
    my_turtle.begin_fill()
    my_turtle.goto(points[1][0], points[1][1])
    my_turtle.goto(points[2][0], points[2][1])
    my_turtle.goto(points[0][0], points[0][1])
    my_turtle.end_fill()


def get_mid(p1, p2):
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)


def sierpinski(points, degree, my_turtle):
    colormap = ["blue", "red", "green", "white", "yellow", "violet", "orange"]
    draw_triangle(points, colormap[degree], my_turtle)
    if degree > 0:
        sierpinski(
            [points[0], get_mid(points[0], points[1]), get_mid(points[0], points[2])],
            degree - 1,
            my_turtle,
        )
        sierpinski(
            [points[1], get_mid(points[0], points[1]), get_mid(points[1], points[2])],
            degree - 1,
            my_turtle,
        )
        sierpinski(
            [points[2], get_mid(points[2], points[1]), get_mid(points[0], points[2])],
            degree - 1,
            my_turtle,
        )


def main():
    my_turtle = turtle.Turtle()
    my_win = turtle.Screen()
    my_points = [[-180, -150], [0, 150], [180, -150]]
    sierpinski(my_points, 5, my_turtle)
    my_win.exitonclick()


main()
```

```python
def factorial(n, depth=0):
    # Print the current "snapshot": n, depth, and call stack
    print(f"{'  ' * depth}Snapshot at depth {depth}: factorial({n})")
    
    # Base case: 1! = 1
    if n == 1:
        print(f"{'  ' * depth}Base case reached: returning 1")
        return 1

    # Recursive step: n! = n * (n-1)!
    result = n * factorial(n - 1, depth + 1)
    
    # Print the snapshot where computation resumes after returning from recursion
    print(f"{'  ' * depth}Returning from depth {depth}: {n}! = {result}")
    
    return result

# Example call
factorial(4)
```

```python
def permute(lst, start=0, depth=0):
    # Print the current "snapshot": the list state and depth
    print(f"{'  ' * depth}Snapshot at depth {depth}: {lst[:start]} + [{lst[start:]}]")
    
    n = len(lst)
    if start == n - 1:
        print(f"{'  ' * depth}Permutation complete: {''.join(lst)}")
        return
    
    for j in range(start, n):
        # Swap to generate a new permutation
        lst[start], lst[j] = lst[j], lst[start]
        permute(lst, start + 1, depth + 1)
        # Swap back to backtrack and restore the list for the next permutation
        lst[start], lst[j] = lst[j], lst[start]

# Example call with a list of characters
permute(list('abc'))
```

```python
def permute_iterative(lst):
    stack = [(lst, 0)]  # Stack of (current list state, current start index)
    results = []  # To store the permutations

    while stack:
        current_lst, start = stack.pop()
        print(current_lst, start)
        if start == len(current_lst) - 1:
            results.append(''.join(current_lst))
        else:
            for j in range(start, len(current_lst)):
                print(start, j)
                # Make a copy of the current list to modify
                next_lst = current_lst[:]
                print('swap: ', next_lst[start], next_lst[j], '=', next_lst[j], next_lst[start])
                
                # Swap to create a new permutation
                next_lst[start], next_lst[j] = next_lst[j], next_lst[start]
                # Push the new state onto the stack
                stack.append((next_lst, start + 1))

    return results

# Example usage
print(permute_iterative(list('abc')))
```

```python
# Top down recursive flow
# - call stack is built first then results are returned

def top_down(n):
    if n == 0:
        return
    print(n)
    top_down(n - 1)

top_down(5)
```

```python
# Bottom up recursive flow
# - call stack is built on return 

def bottom_up(n):
    if n == 0:
        return
    bottom_up(n - 1)
    print(n)

bottom_up(5)
```

```python
def permute(chars, path, used):
    
    if len(path) == len(chars):
        print(path)
        return

    for i in range(len(chars)):
        
        if used[i]:
            continue
            
        path.append(chars[i])
        used[i] = True

        permute(chars, path, used)

        path.pop()
        used[i] = False

        
```

```python
def permute(chars, path, used, depth=0):
    indent = "  " * depth  # Indentation to visualize depth of recursion

    if len(path) == len(chars):
        print(f"{indent}✅ Base case hit: {path}")
        return

    print(f"{indent}Entering level {depth} with path={path}, used={used}")

    for i in range(len(chars)):
        if used[i]:
            print(f"{indent}  Skipping used char '{chars[i]}' at i={i}")
            continue

        # Choose
        print(f"{indent}  Choosing '{chars[i]}' at i={i}")
        path.append(chars[i])
        used[i] = True

        # Recurse
        permute(chars, path, used, depth + 1)

        # Backtrack
        popped = path.pop()
        used[i] = False
        print(f"{indent}  Backtracking: removed '{popped}', path now {path}, used={used}")

    print(f"{indent}Leaving level {depth} with path={path}")
```

```python
permute(['A', 'B'], [], [False, False])
```

```python
def permute(s, prefix=''):
    if len(s) == 0:
        print(prefix)
    else:
        for i in range(len(s)):
            permute(s[:i] + s[i+1:], prefix + s[i])

permute('ABC')
```
