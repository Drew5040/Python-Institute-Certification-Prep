```python

# Reverse list loop (in-place)
def reverseList(lst):
    i=0
    j=len(lst)-1
    
    while i < j:
        lst[i], lst[j] = lst[j], lst[i]
        i+=1
        j-=1
    return lst


A = [1, 2, 3, 4, 5, 6, 7] 
print(A) 
reverseList(A)
```

```python
# Reverse list recursive

def reverseListRecursive(l, n, m):
    if n >= m:
        return
    l[n], l[m] = l[m], l[n]
    return reverseListRecursive(l, n+1, m-1)
    
lst = reverseListRecursive(A, 0, len(A)-1)   
print(A)
```

```python
# Reverse list (stack)

def revListStack(arr):
    stack = []
    
    for num in arr:
        stack.append(num)
    
    print(stack)
    for i in range(len(stack)):
        arr[i] = stack.pop()
        
    return arr

revListStack(A)
        
```

```python
# Counter-clockwise rotation
def rotateLst(arr):
    first_elem = arr[0]
    for i in range(len(arr)-1):
        arr[i] = arr[i+1]
    arr[-1] = first_elem
    return arr

A = [1, 2, 3, 4, 5, 6, 7]
rotateLst(A)
        
```
