## Tabulation -Vs- Memoization

```python
# Tabulation: Bottom Up
# Memoization: Top Down
```

```python
### > Tabulation: Bottom Up

def factorial(n):

# initialize dp array with 0s
    dp = [0] * (n+1)

# base case
    dp[0] = 1

# calculate factorials up to n
    for i in range(1, n+1):
        dp[i] = dp[i-1] * i
        
    return dp
```
